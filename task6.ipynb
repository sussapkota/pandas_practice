```python
import pandas as pd
```

```python
df = pd.read_csv('Attributes.csv')
```

```python
df.head(3)
```

```python
def standardize_columns(col_names):
    final = []
    for item in col_names:
        item = item.lower()
        item = item.split(' ')
        item = "_".join(item)
        final.append(item)
    return final
```

```python
df.columns = standardize_columns(df.columns)
```

```python
df.columns
```

```python
# combining multiple columns into one
```

```python
df['image'] = df[df.columns[18:]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)
```

```python
df= df.drop(['default_image', 'front_image', 'front_image',
       'side_image', 'back_image', 'detail_angle', 'look_shot_image',
       'additional_image_1', 'additional_image_2'],axis=1)
```

```python
list(df.head(1)['image'])
```

```python
DESIRED_TABLE = ['description','material_&_care', 'article_type', 'fabric', 'fit', 'neck', 'pattern','type',
                 'length', 'country_of_origin', 'waist_rise', 'wash_care','sleeve_length']
def create_table(row):
    row = dict(row)
    
    first = '<table border="0">'
    
    for k,v in row.items():
        if str(v) !='nan' and k in DESIRED_TABLE:
            first = first + '<tr><td><strong>' + str(k)+'</strong></td></tr>'
            first = first + '<tr><td><p>' + str(v)+ '</p></td><tr>'
    last = '</table>'
    
    return first+last
    
    
    
```

```python
df['attributes']=df.apply(create_table,1)
```

```python
list(df.head(2)['attributes'])
```

```python
# now to get price
```

```python
price_df = pd.read_csv('Invoice.csv')
```

```python
price_df.head()
```

```python
cols = price_df.columns
cols
```

```python
price_df = price_df.drop(['HS Code', 'DESCRIPTION OF GOODS', 'SKU Code', 'Brand',
       'Dispatch Qty', 'TOTAL VALUE\nIN INR'],axis=1)
```

```python
price_df.columns = ['style_id','PRICE PER UNIT\nIN INR']
```

```python
result = pd.merge(df, price_df, how="inner", on=["style_id"])
```

```python
result.head()
```

```python
list(df.head(1)['attributes'])
```

```python
result = result.apply(lambda x: x*1.6 if x.name == 'PRICE PER UNIT\nIN INR' else x)
```

```python
df = result.rename(columns={'PRICE PER UNIT\nIN INR': 'price'})
```

```python
df.head(2)
```

```python
# now lets get quantity now
```

```python
quantity_df = pd.read_csv('Packing-List.csv')
```

```python
quantity_df.head()
```

```python
quantity_df = quantity_df.groupby(['Style Id'],as_index=False)['Ship Quantity'].sum()
```

```python
quantity_df.head()
```

```python
quantity_df = quantity_df.rename(columns={'Style Id': 'style_id','Ship Quantity':'quantity'})
```

```python
result = pd.merge(df, quantity_df, how="inner", on=["style_id"])
```

```python
result[['style_id','quantity']].head()
```

```python
result.columns
```

```python
# discarding the attribute column
result.to_csv('final.csv',columns=['style_id', 'gender', 'brand', 'style_name', 'color', 'description',
       'material_&_care', 'article_type', 'fabric', 'fit', 'neck', 'pattern',
       'type', 'length', 'country_of_origin', 'waist_rise', 'wash_care',
       'sleeve_length', 'image', 'price', 'quantity'],index=False)
```
