```python
import pandas as pd

from pandas import DataFrame
```

```python

class SusonDataframe(DataFrame):

    def melt(
            self,
            id_vars=None,
            value_vars=None,
            var_name=None,
            value_name="value",
            col_level: None = None,
            ignore_index: bool = True,
    ) -> DataFrame:
        #  Implementation goes here
        self.df_list = list()

        self.apply(self.process, 1, args=(id_vars, value_name))

        return DataFrame(data=self.df_list)

    def process(self, row, id_vars, value_name):
        test = dict(row)
        for key, value in test.items():
            output_dict = dict()
            if key not in id_vars:
                for val in id_vars:
                    output_dict.update({val: row[val]})
                output_dict.update({'variable': key})
                output_dict.update({value_name: value})

                self.df_list.append(output_dict)

```

```python

df = pd.read_csv('DataSetThree.csv')
df.head()
```

```python
df = SusonDataframe(df)
df.melt(id_vars=['employee_id', 'name'], value_name="Count")
```
